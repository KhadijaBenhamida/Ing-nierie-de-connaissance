<h3 style="color:blue;">Exercice 1 : Réseau sémantique – Chauve-souris</h3>

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
```

```python

noeuds = ["Animal", "Oiseau", "Mammifère",
          "Chauve-souris", "Voler", "Pondre", "Allaiter"]
G.add_nodes_from(noeuds)

G.add_edge("Oiseau", "Animal", relation="est un")
G.add_edge("Mammifère", "Animal", relation="est un")
G.add_edge("Chauve-souris", "Oiseau", relation="est un")
G.add_edge("Chauve-souris", "Mammifère", relation="est un")

G.add_edge("Oiseau", "Voler", relation="peut")
G.add_edge("Oiseau", "Pondre", relation="peut")
G.add_edge("Mammifère", "Allaiter", relation="peut")

G.add_edge("Chauve-souris", "Pondre", relation="exclusion")
```

```python

pos = {
    "Animal": (0, 3),
    "Oiseau": (-2, 2),
    "Mammifère": (2, 2),
    "Chauve-souris": (0, 1),
    "Voler": (-3, 0),
    "Pondre": (-1.5, 0),
    "Allaiter": (2.5, 0)
}
```

```python
couleurs = []
for n in G.nodes():
    if n in ["Animal", "Oiseau", "Mammifère", "Chauve-souris"]:
        couleurs.append("#cfe8ff")
    else:
        couleurs.append("#8fbce6")

nx.draw_networkx_nodes(
    G, pos,
    node_color=couleurs,
    node_size=2600,
    edgecolors="#3b82f6",
    linewidths=2
)

nx.draw_networkx_labels(
    G, pos,
    font_size=10,
    font_weight="bold",
    font_color="#0f172a"
)

nx.draw_networkx_edges(
    G, pos,
    arrows=True,
    arrowstyle='-|>',
    arrowsize=20,
    width=2,
    edge_color="#3b82f6"
)

edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color="#1e3a8a",
    font_size=9,
    font_weight="bold"
)

plt.title(
    " Réseau sémantique : Chauve-souris, Oiseau, Mammifère ",
    fontsize=13,
    fontweight="bold",
    color="#1e3a8a"
)

plt.axis("off")
plt.tight_layout()
plt.show()
```


<h4 style="color:blue;"> 2️-Héritage :</h4>

La chauve-souris hérite des propriétés :

de Oiseau → “Peut voler”

de Mammifère → “Allaite ses petits”

et possède aussi sa propre propriété : “Ne pond pas d’œufs”.

<h4 style="color:blue;">  3️-Conflits :</h4>

Un conflit sémantique apparaît : -La chauve-souris est un “oiseau” (donc censée pondre des œufs)

-Mais elle “ne pond pas d’œufs” → contradiction biologique.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
```

```python

categories = ["Personne", "Étudiant", "Enseignant", "Professeur", "Chercheur"]
proprietes = ["Domaine", "Cours", "Âge"]

G.add_nodes_from(categories + proprietes)

G.add_edge("Étudiant", "Personne", relation="est-un")
G.add_edge("Enseignant", "Personne", relation="est-un")
G.add_edge("Professeur", "Enseignant", relation="est-un")
G.add_edge("Professeur", "Chercheur", relation="est-un")

G.add_edge("Personne", "Âge", relation="possède")
G.add_edge("Étudiant", "Domaine", relation="étudie")
G.add_edge("Enseignant", "Domaine", relation="enseigne")
G.add_edge("Cours", "Domaine", relation="D.étudié = D.enseigné")
```

```python

pos = {
    "Personne": (0, 3),
    "Étudiant": (-2, 2),
    "Enseignant": (2, 2),
    "Professeur": (2, 1),
    "Chercheur": (4, 1),
    "Domaine": (0, 1),
    "Cours": (0, 0),
    "Âge": (-3, 3)
}
```

La règle stipulant que « le domaine enseigné doit être le même que le domaine étudié pour un cours donné » représente une relation d’égalité, considérée comme une contrainte d’intégrité dans un réseau sémantique.

Elle oblige à maintenir une adéquation exacte entre les informations associées à un même cours.

Cette règle relève des contraintes de cohérence, puisqu’elle assure l’alignement et la conformité des données décrivant le cours.

<h3 style="color:blue;">Exercice 4 </h3>


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
```

```python

nodes = ["ÊtreVivant", "Animal", "Chien", "Chat", "Os", "Lait", "Nourriture"]
G.add_nodes_from(nodes)


G.add_edge("Chien", "Animal", relation="est-un")
G.add_edge("Chat", "Animal", relation="est-un")
G.add_edge("Animal", "ÊtreVivant", relation="est-un")
G.add_edge("Chien", "Os", relation="lié-à")
G.add_edge("Chat", "Lait", relation="lié-à")
G.add_edge("Os", "Nourriture", relation="type")
```

```python
pos = {
    "ÊtreVivant": (0, 4),
    "Animal": (0, 3),
    "Chien": (-2, 2),
    "Chat": (2, 2),
    "Os": (-2, 1),
    "Lait": (2, 1),
    "Nourriture": (0, 0)
}
```

```python
activation = {"Chien": 1.0}
facteur = 0.5
niveau = 3

for _ in range(niveau):
    new_activation = activation.copy()
    for node, value in activation.items():
        for neighbor in G.successors(node):
            propagated = value * facteur
            if neighbor not in new_activation or propagated > new_activation[neighbor]:
                new_activation[neighbor] = propagated
    activation = new_activation

max_act = max(activation.values())
```

```python
colors = []

for node in G.nodes():
    val = activation.get(node, 0)
    intensity = val / max_act if max_act else 0

    r = (1 - intensity) * (255/255) + intensity * (230/255)
    g = (1 - intensity) * (179/255) + intensity * (179/255)
    b = (1 - intensity) * (217/255) + intensity * (255/255)

    colors.append((r, g, b))

plt.figure(figsize=(8, 8))
```

```python
nx.draw_networkx_nodes(
    G, pos,
    node_color=colors,
    node_size=3000,
    edgecolors="#3b82f6",
    linewidths=2
)

nx.draw_networkx_edges(
    G, pos,
    arrowstyle='-|>',
    arrowsize=25,
    width=2,
    edge_color="#a4c8f5"
)

nx.draw_networkx_labels(
    G, pos,
    font_size=11,
    font_weight='bold',
    font_color="#1e3a8a"
)

edge_labels = nx.get_edge_attributes(G, "relation")
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color="#1e3a8a",
    font_size=9,
    font_weight="bold"
)

plt.title(
    "Propagation d’activation (nœud initial = Chien)",
    fontsize=14,
    fontweight='bold',
    color="#1e3a8a"
)

plt.axis('off')
plt.tight_layout()
plt.show()

print("Niveau d'activation après propagation :")
for node, val in activation.items():
    print(f"{node:<12} → activation = {val:.3f}")
```

<h3 style="color:blue;">Exercice 6 </h3>


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
```

```python
nodes = ["Personne", "Étudiant", "Travailleur", "Ali"]
G.add_nodes_from(nodes)

G.add_edge("Étudiant", "Personne", relation="est-un")
G.add_edge("Étudiant", "Travailleur", relation="n’est-pas")
G.add_edge("Ali", "Étudiant", relation="est-un")
```

```python

pos = {
    "Personne": (0, 3),
    "Étudiant": (0, 2),
    "Travailleur": (-2, 2),
    "Ali": (0, 1)
}
```

```python
node_colors = []
for node in G.nodes():
    if node == "Ali":
        node_colors.append("#a4c8f5")
    else:
        node_colors.append("#d6eaff")

plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(
    G, pos,
    node_color=node_colors,
    node_size=2500,
    edgecolors="#3b82f6",
    linewidths=2
)

nx.draw_networkx_labels(
    G, pos,
    font_size=11,
    font_weight='bold',
    font_color="#1e3a8a"
)

for u, v, data in G.edges(data=True):
    relation = data['relation']

    if relation == "sorte-de":
        style = "solid"
        color = "#3b82f6"
    elif relation == "n’est-pas":
        style = "dotted"
        color = "#1e3a8a"
    else:
        style = "dashed"
        color = "#5a8fdc"

    nx.draw_networkx_edges(
        G, pos,
        edgelist=[(u, v)],
        style=style,
        width=2,
        arrowstyle='-|>',
        arrowsize=25,
        edge_color=color
    )

edge_labels = nx.get_edge_attributes(G, "relation")
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color="#1e3a8a",
    font_size=9,
    font_weight="bold"
)

plt.title(
    "Réseau sémantique : Personne, Étudiant, Travailleur & Ali",
    fontsize=14,
    fontweight='bold',
    color="#1e3a8a"
)

plt.axis('off')
plt.tight_layout()
plt.show()
```
